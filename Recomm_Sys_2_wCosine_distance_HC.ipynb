```python
'''
# Spotify  provides wide range of fatures about songs such as : Popularity, 
 Explicit, Danceability, Energy, Key, Loudness, Mode, Speechiness,
 Acousticness, Instrumentalness, Liveness, Valence,Tempo

Most of these features are already numeric or with a little bit of data
manipulation they can be converted into numeric data. Numeric features
allows calculating cosine similarity between songs. I found the cosine
distance between songs and offered a recommendation system based on
cosine distance. 
'''

```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("music_data.csv")
```

```python
df.drop(['Unnamed: 0','Album ID', 'Track ID', 'External URLs','Release Date','Duration (ms)'], axis =1 , inplace=True)
```

```python
df['Explicit'].replace({True: 1, False: 0}, inplace = True) 
```

```python
df_c = df[['Popularity','Explicit', 'Danceability', 'Energy', 'Key', 'Loudness', 'Mode',\
       'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness',\
       'Valence', 'Tempo']]
similarity = cosine_similarity(df_c)
```

```python
def recommend_track(track_name) :
    track_list = df[df['Track Name'].str.contains(track_name)]
    #print(track_list)
    if len(track_list):
        track_idx = track_list.index[0]
        indexes = sorted(list(enumerate(similarity[track_idx])),reverse = True,   key = lambda x:x[1])[1:11]

        for i in indexes :
            track_indx, similarity_val = i;
            print("similarity: {} \t  {}".format(similarity_val, df['Track Name'].iloc[track_indx]) )
    else:
        return ('no tracks found in database')
```

```python
recommend_track('Life Changes')
```
